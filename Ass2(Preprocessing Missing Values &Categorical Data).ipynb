```python
#(4)imputer
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy="most_frequent")
col=data[["Location"]]
imputed_values=imp.fit_transform(col)
new_values=pd.DataFrame(imputed_values)
new_values
```

```python
new_values.columns=col.columns
new_values
```

```python
data.Location=new_values.Location
```
