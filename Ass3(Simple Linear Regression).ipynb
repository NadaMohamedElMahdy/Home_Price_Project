```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("London.csv")
data
```

```python
#simple"one var"linear regression
x_train=data.loc[:2784,["Area in sq ft"]]
y_train=data.loc[:2784,["Price"]]
x_test=data.loc[2785:,["Area in sq ft"]]
y_test=data.loc[2785:,["Price"]]
y_test
```

```python
#visulization
plt.title("rel between x_train and y_train")
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.scatter(x_train,y_train,color="green")
plt.show()
```

```python
#train
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#predict
y_predict=model.predict(x_test)
y_predict
```

```python
#validation'error'
import sklearn.metrics as mc
```

```python
dir(mc)
```

```python
mc.mean_absolute_error(y_test,y_predict)
```

```python
plt.scatter(data["Area in sq ft"],data["Price"])
plt.plot(x_test,y_predict,color="r")
```
