```python
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("London.csv")
data
```

```python
data.drop("Index",axis=1,inplace=True)
data
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
x=data[["Area in sq ft","No. of Bedrooms","No. of Bathrooms","No. of Receptions"]]
y=data.Price
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
x_train
```

```python
#visulization
sns.pairplot(data)
```

```python
sns.pairplot(x_vars=["Area in sq ft","No. of Bedrooms","No. of Bathrooms","No. of Receptions"],y_vars="Price",data=data,kind="scatter",diag_kind=None)
```

```python
#multiple_linear_regression
#(1)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#train
model.fit(x_train,y_train)
```

```python
#test
y_predict=model.predict(x_test)
y_predict
```

```python
plt.scatter(y_test,y_predict)
```

```python
y_test
```

```python
#validtion
import sklearn.metrics as mc
dir(mc)
```

```python
mc.mean_squared_error(y_test,y_predict)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#rename columns
data.rename({"Area in sq ft":"Area_sq_ft","No. of Bedrooms":"Num_of_Bedrooms","No. of Bathrooms":"Num_of_Bathrooms","No. of Receptions":"Num_of_Receptions"},axis=1,inplace=True)
data
```

```python
#statsmodels
import statsmodels.formula.api as sm
stats_model=sm.ols(formula='Price ~ Area_sq_ft + Num_of_Bedrooms + Num_of_Bathrooms + Num_of_Receptions',data=data).fit()
```

```python
y_stats_predict=stats_model.predict({'Area_sq_ft':1000,'Num_of_Bedrooms':3 ,'Num_of_Bathrooms':2,'Num_of_Receptions':2})
y_stats_predict
```
